```python

train_datagen = ImageDataGenerator(rescale=1./255)
train_batches = train_datagen.flow_from_directory('IMG2/',
        class_mode = 'input', batch_size=BATCH_SIZE,target_size=(156,108))
```

```python
x = Input(shape=(156, 108, 3))
    # Encoder
e_conv1 = Conv2D(64, (3, 3), activation='relu', padding='same')(x)
pool1 = MaxPooling2D((2, 2), padding='same')(e_conv1)
batchnorm_1 = BatchNormalization()(pool1)
e_conv2 = Conv2D(48, (3, 3), activation='relu', padding='same')(batchnorm_1)
pool2 = MaxPooling2D((2, 2), padding='same')(e_conv2)
batchnorm_2 = BatchNormalization()(pool2)
e_conv3 =  Conv2D(24, (3, 3), activation='relu', padding='same')(batchnorm_2)
pool3 = MaxPooling2D((2, 2), padding='same')(e_conv3)
batchnorm_3 = BatchNormalization()(pool3)
e_conv4 = Conv2D(12, (3, 3), activation='relu', padding='same')(batchnorm_3)
encoded = MaxPooling2D((2, 2), padding='same')(e_conv4)
    
latentSize = (7, 10, 12)
# Decoder
direct_input = Input(shape=latentSize)
d_conv1 = Conv2D(12, (3, 3), activation='relu', padding='same')(direct_input)
up1 = UpSampling2D((2, 2))(d_conv1)
d_conv2 = Conv2D(24, (3, 3), activation='relu', padding='same')(up1)
up2 = UpSampling2D((2, 2))(d_conv2)
d_conv3 = Conv2D(48, (3, 3), activation='relu', padding='same')(up2)
up3 = UpSampling2D((2, 2))(d_conv3)
d_conv4= Conv2D(64, (3, 3), activation='relu')(up3)
up4 = UpSampling2D((2, 2))(d_conv4)
decoded = Conv2D(3, (3, 3), activation='sigmoid', padding='same')(up4)
   
encoder = Model(x, encoded)
decoder = Model(direct_input, decoded)
autoencoder = Model(x, decoder(encoded))
autoencoder.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
images = next(iter(train_batches))[0]
for i in range(5):

    plt.figure(figsize=(10,15))
    plt.subplot(1,3,1)
    
    # ORIGINAL IMAGE
    orig = images[i,:,:,:].reshape((-1,156,108,3))
    img = Image.fromarray( (255*orig).astype('uint8').reshape((156,108,3)))
    plt.title('Original')
    plt.imshow(img)
    
    
    latent_img = encoder.predict(orig)
    mx = np.max( latent_img[0] )
    mn = np.min( latent_img[0] )
    latent_flat = ((latent_img[0] - mn) * 255/(mx - mn)).flatten(order='F')
    img = Image.fromarray( latent_flat[:2025].astype('uint8').reshape((28,30)), mode='L') 
    plt.subplot(1,3,2)
    plt.title('Latent')
    plt.xlim((-10,55))
    plt.ylim((-10,55))
    plt.axis('off')
    plt.imshow(img)
    
    
    latent_img = autoencoder.predict(orig)
    img = Image.fromarray( (255*latent_img[0]).astype('uint8').reshape((156,108,3)))
    plt.subplot(1,3,3)
    plt.title('Reconstructed')
    plt.imshow(img)
```
